```python
import numpy as np
```

1. Flatten a given nested list

```python
my_list = [[1, 2, 3], [4, 5, 6]]

flat_list = []
for sublist in my_list:
    for num in sublist:
        flat_list.append(num)

print(flat_list)
```

```python
arr = np.array(my_list, dtype=object)
flat_arr = arr.flatten()
print (flat_arr)
```

2. Write a program to add a border of zeros around the existing array.

```python
import numpy as np

# Create NumPy arrays filled with ones
ones_arr = np.ones((4,4))

print("Transformed array:")
transformed_array = np.pad(ones_arr, pad_width=1, mode='constant', constant_values=0)
print(transformed_array)
```

```python
import numpy as np

# Create NumPy arrays filled with ones
ones_arr = np.ones((4,4))

print("Transformed array:")
transformed_array = np.pad(ones_arr, pad_width=1, mode='edge')
print(transformed_array)
```

#Data Visualization

1. Matplotlib

```python
import matplotlib.pyplot as plt

weekdays = ['sun', 'mon', 'tue']
sale = [12, 13, 11]
plt.bar(weekdays, sale)
plt.show()
```

```python
import seaborn as sns

weekdays = ['sun', 'mon', 'tue']
sale = [12, 13, 11]
sns.barplot(x=weekdays, y=sale)
```

```python
import plotly.express as px

weekdays = ['sun', 'mon', 'tue']
sale = [12, 13, 11]
fig = px.bar(x=weekdays, y=sale)
fig.show()
```
